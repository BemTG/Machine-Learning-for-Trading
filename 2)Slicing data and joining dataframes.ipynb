```python
import pandas as pd 
import matplotlib.pyplot as plt
import pandas_datareader as data
```

```python
df=data.DataReader(['AAPL','IBM','TSLA','AMZN','JD'],'yahoo',start='2012-01-01',end='2015-01-01')
df
```

In order to get specific data from the dataframe, you can slice the data 
using  df.loc . 

This allows for specific rows from specific columns to be retrieved without 
needing to print the whole dataframe and find data.

```python
#create a new df using only the adj close for all stocks and drop NaN values
df=df['Adj Close']
df=df.dropna()

#Use .loc to get data from 2014/06/06 to 2014/09/09 for only IBM and AMZN
df=df.loc['2014-06-06':'2014-09-09',['IBM','AMZN','TSLA']]
df
```

```python
#plot
def plot(df):
    df=df.plot()
    return plt.show()

plot(df)
```

## Joining dataframes 

```python
def new_dates():
    #create a date range
    start_date='2015-01-01'
    end_date='2020-01-01'
    dates=pd.date_range(start_date,end_date)
    
    #create a dataframe and put the date range as the index col in the df
    df0=pd.DataFrame(index=dates)
    
    #Read the stock csv file, use the neccessary columns only and set date 
    #as the index col
    df1=pd.read_csv('AAPL.csv',parse_dates=True, 
                    usecols=['Date','Adj Close'],
                   na_values=['nan'])
    df1=df1.set_index('Date')
    
    
    #join the two dataframes; dataframes will join based on the index col
    df0=df0.join(df1, how='inner')
    #drop all the non values
    df0=df0.dropna()
    print(df0)
    
new_dates()
```
