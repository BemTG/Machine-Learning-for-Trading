```python
import pandas as pd
import numpy as np
import pandas_datareader as dt
import matplotlib.pyplot as plt
```

```python
df=dt.DataReader('^GSPC', 'yahoo', start='2020/01/01')
df
```

```python
df=df[['Close']]
df
```

```python


def upper_bollinger_bands(df,window):
    rm=df.rolling(window).mean()
    std=df.rolling(window).std()
    upper= rm+ std*2
    return upper

def lower_bollinger_bands(df,window):
    rm=df.rolling(window).mean()
    std=df.rolling(window).std()
    lower= rm- std*2
    return lower

#df.rolling(2).sum()
    
```

```python
df['Upper B.B']=upper_bollinger_bands(df,10)
df
```

```python
df['Lower B.B']=lower_bollinger_bands(df[['Close']],10)
df
```

```python
df.plot()
plt.show()
```

### Cumalitive returns

```python
df2=dt.DataReader('^GSPC', 'yahoo', start='01/01/2018', end='01/01/2020')
df2=df2[['Close']]
df2
```

```python
df2['PCT change']=df2.pct_change()
df2['PCT change sum']=df2[['PCT change']].cumsum()+1
df2


```

```python
df2['PCT change sum'].plot()
plt.show()
df2['Close'].plot()
plt.show()
```
