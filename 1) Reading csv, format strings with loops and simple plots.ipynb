# Plotting

```python
import matplotlib.pyplot as plt
```

Pandas allows us to plot data using .plot() if the dataframe has a an index 
already. 

matplotlib.pyplot library allows us to visualise the plot through plt.show()

```python

def plot_graph(symbol):
    df=pd.read_csv('{}.csv'.format(symbol))
    df['Close'].plot()
    return plt.show()

for symbol in symbols:
    plot_graph(symbol)

    
```
