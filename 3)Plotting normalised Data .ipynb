```python
import pandas as pd 
import matplotlib.pyplot as plt
import pandas_datareader as data
```

```python
df=data.DataReader(['AAPL','IBM','TSLA','AMZN','JD'],'yahoo',start='2012-01-01',end='2015-01-01')
df
```

```python
#create a new df using only the adj close for all stocks and drop NaN values
df=df['Adj Close']
df=df.dropna()

#Use .loc to get data from 2014/06/06 to 2014/09/09 for only IBM and AMZN
df=df.loc['2014-06-06':'2014-09-09',['IBM','AMZN','TSLA']]
df
```

```python
#plot
bamia='bamis'
def plot(df,title='Stock prices'):
    ax= df.plot(title=title,fontsize=10)
    
    #add the y and x label
    ax.set_xlabel('Date')
    ax.set_ylabel('Price')
    
    return plt.show()

plot(df)
```

#### Normalising the data

We can normalise the dataframe by deviding the whole dataframe by the initial
row of data. 

This will illustrate where the stock price is in relation to its firs price.
Therefore all the stocks in the dataframe will be measured in comparison of the starting date.

This normalisation method allows us to see how much would a stock have gained/lost based on percentage at a given time starting from the starting date.
This makes it easier to visualise and compare different stocks preformance in relation to their starting date

```python


df2=df/df.loc['2014-06-06']
df2
```

```python
plot(df2)
```
